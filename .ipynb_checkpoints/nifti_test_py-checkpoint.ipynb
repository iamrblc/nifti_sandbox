 # Handling nifti files in Python

## Necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import nibabel as nib
```

## Reading nifti files

```python
nifti_path = 'test_nifti_1.nii'
img = nib.load(nifti_path)
```

```python
img.shape
```

```python
img.get_data_dtype()
```

```python
img.header
```

```python
data = img.get_fdata()
```

```python
plt.imshow(data[:, :, 30], cmap='inferno')
plt.show()
```

```python
from nilearn import plotting
```

```python
# Plot a glass brain projection of the image
plotting.plot_glass_brain(img)
# Or plot an individual slice
plotting.plot_epi(img, display_mode='z', cut_coords=5)
plotting.show()
```
